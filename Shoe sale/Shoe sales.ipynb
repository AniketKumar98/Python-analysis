```python
#Sales in US
d=df[df['Country']=='United States']
d
```

```python
#Male customers in US 
dff=d[d['Gender']=='Male']
dff
```

```python
#Male customer if size 10 in US
xf=dff[dff['Size (US)']==10]
xf
```

```python
#Splitting date and time
x=xf['Date'].apply(lambda x: str(x).split('-'))
x
```

```python
#Number of sales in July,2015 of size 10
sum(x.apply(lambda a: a[0]=='2015' and a[1]=='07'))
```
